```python
def checkifseated(LHip,RHip,LKnee,RKnee,normalization_factor):
  if 0<= abs(LKnee[1] - LHip[1])<= 55 or 0<= abs(RKnee[1] - RHip[1])<= 55 :
    print('seated') 
    return True
  else:
    print('standing/walking')
    return False 

def checkifspread(LShoulder,RShoulder, LWrist, RWrist, LElbow, RElbow):
  if LShoulder[0] < LElbow[0] <LWrist[0] or RShoulder[0] > RElbow[0] > RWrist[0]:
    print('arm/s spread while seated')
    return True
  else:
    print('arm/s not spread')
    return False

def checkstance(LShoulder, RShoulder, LAnkle, RAnkle):
  shoulder_dist=math.sqrt(((LShoulder[0]-RShoulder[0])**2) +((LShoulder[1]-RShoulder[1])**2))
  feet_dist = math.sqrt(((LAnkle[0] - RAnkle[0])**2) + ((LAnkle[1]-RAnkle[1])**2))
  if (shoulder_dist + (shoulder_dist*0.1)) > feet_dist > (shoulder_dist - (shoulder_dist*0.1)):
    print('confident stance')
    return True
  else:
    print('neutral stance')
    return False

def checkifkneesapart(LKnee,RKnee,LShoulder,RShoulder):
  if (LKnee[0] > LShoulder[0] and RKnee[0] < RShoulder[0]):
    print('sitting with knees apart')
    return True
  else:
    print('knees not spread')
    return False

def handcoverseye(LEye, REye, LWrist, RWrist):
  if ((0<=abs(LEye[1]-LWrist[1])<=30) and (0<=abs(LEye[0]-LWrist[0])<=30)) or ((0<=abs(REye[1]-RWrist[1])<=30)and(0<=abs(REye[0]-RWrist[0])<=30)):
    print("Hands covering Eyes")
  else:
    print("Hands not covering eyes")

def handscovernose(LWrist, RWrist, Nose):
 
  #if LWrist[0] > Nose[0] > (LWrist[0] - (LWrist[0]* 0.023)) and RWrist[0] < Nose[0] < (RWrist[0] + (RWrist[0]* 0.023)):
    #if Nose[1] < LWrist[1] < (Nose[1] + (Nose[1]*0.29)) and Nose[1] < RWrist[1] < (Nose[1] + (Nose[1]*0.29)):
    print(LWrist)
    print(RWrist)
    print(Nose)
    if 0<=(LWrist[0]-Nose[0]) <=60 and 0<=(LWrist[1]-Nose[1])<=60 and 0<= (Nose[0]-RWrist[0]) <=60 and 0<=(RWrist[1]-Nose[1])<=60:
      print('Hands are covering nose')
      return True
    else:
      print('hands are not covering nose')
      return False

  

def earpull(LEar, REar, LWrist, RWrist):
  
  #if LEar[0] < LWrist[0] < (LEar[0] + (LEar[0]*0.04)) or REar[0]> LWrist[0] > (REar[0] - (REar[0] * 0.04)):
    #if LEar[1] < LWrist[1] < (LEar[1] + (LEar[0]* 0.8 )) or  REar[1] < LWrist[1] < (REar[1] + (RLEar[1]*0.8 )):
      #print('Hand is pulling lobe')
      #return
  print(LEar)
  print(REar)
  print(LWrist)
  print(RWrist)
  if 0<= (LEar[1]-LWrist[1])<=40 or 0<=(REar[1]-RWrist[1])<=40:
    print("Hand is pulling lobe")
    return

  else:
   print('Hand not on earlobe')
   return

def hipsout(LHip, RHip, LShoulder, RShoulder):
  
  if LHip[0] > LShoulder[0] or RHip[0] < RShoulder[0]:
    print('Hips out')
    return True
  else:
    print('Hips not out')
    return False

def checkspreadelbows(LWrist, RWrist, LElbow, RElbow, LShoulder, RShoulder,normalization_factor):
 
  #if LShoulder[0] + (LShoulder[0]* 0.3)> LElbow[0] > LShoulder[0]  or RShoulder[0] - (RShoulder[0]*0.3) < RElbow[0] < RShoulder[0]: 
  if LElbow[0]>LShoulder[0] and RElbow[0]<RShoulder[0]:
    #if LElbow[1] + (LElbow[1]* 0.06) > RWrist[1] > LElbow[1] - (LElbow[1]*0.06) and RElbow[1] + (RElbow[1]* 0.06) > LWrist[1] > RElbow[1] - (RElbow[1]*0.06):
    if 1>= (LElbow[1]-LWrist[1])/normalization_factor >= 0 and 1>= (RElbow[1]-RWrist[1])/normalization_factor >= 0 :
      print('elbows spread')
      return True
    else :
      print('elbows not spread')
      return False
  else :
    print('elbows not spread')
    return False

import math
import json
%matplotlib inline
from matplotlib import image
from matplotlib import pyplot as plt

f = open('/content/gdrive/MyDrive/finalproject/Major/examples/res_single/results.json',)
Data = json.load(f)
count = 0
for data in Data:
  image_name = data['image_id']
  data = data['keypoints']
  image_data = image.imread(f'/content/gdrive/MyDrive/finalproject/Major/examples/demo_single/{image_name}')
  coordinates = []
  scores = []
  c = 0
  temp_coordinates = []
  for temp in data:
      if c%3 == 0:
          temp_coordinates.append(temp)
      elif c%3 == 1:
          temp_coordinates.append(temp)
        #print(temp_coordinates)
          coordinates.append(temp_coordinates)
      else:
          temp_coordinates = []
          scores.append(temp)
        
      c=c+1
    
#print(coordinates)
#print(scores)
  f.close()
  

  
  Nose = coordinates[0]
  LEye = coordinates[1]
  REye = coordinates[2]
  LEar = coordinates[3]
  REar = coordinates[4]
  LShoulder = coordinates[5]
  RShoulder = coordinates[6]
  LElbow = coordinates[7]
  RElbow = coordinates[8]
  LWrist = coordinates[9]
  RWrist = coordinates[10]
  LHip = coordinates[11]
  RHip = coordinates[12]
  LKnee = coordinates[13]
  RKnee =coordinates[14]
  LAnkle = coordinates[15]
  RAnkle = coordinates[16]
 
  #dist1 = math.sqrt( ((LEye[0] - RAnkle[0])**2) + ((LEye[1] - RAnkle[1])**2) )
  #dist2 = math.sqrt( ((REye[0]- LAnkle[0])**2) +  ((REye[1] - LAnkle[1])**2) )
  normalization_factor = RAnkle[1]-Nose[1]
  '''
  print(dist1, dist2)

  if dist1 >dist2:
    great_dist = dist1
    great_dist = (1/great_dist)
  else:
    great_dist = dist2
    great_dist = (1/great_dist)
  '''

 

  plt.plot( Nose[0], Nose[1], marker = 'v', color ="white")
  plt.text( Nose[0], Nose[1], 'Nose')
  plt.plot( LEye[0], LEye[1], marker = 'v', color ="white")
  plt.text( LEye[0], LEye[1], 'LEye')
  plt.plot( REye[0], REye[1], marker = 'v', color ="white")
  plt.text( REye[0], REye[1], 'REye')
  plt.plot( LEar[0], LEar[1], marker = 'v', color ="white")
  plt.text( LEar[0], LEar[1], 'LEar')
  plt.plot( REar[0], REar[1], marker = 'v', color ="white")
  plt.text( REar[0], REar[1], 'REar')
  plt.plot( LShoulder[0], LShoulder[1], marker = 'v', color ="white")
  plt.text( LShoulder[0], LShoulder[1], 'LShoulder')
  plt.plot( RShoulder[0], RShoulder[1], marker = 'v', color ="white")
  plt.text( RShoulder[0], RShoulder[1], 'RShoulder')
  plt.plot( LElbow[0], LElbow[1], marker = 'v', color ="white")
  plt.text( LElbow[0], LElbow[1], 'LElbow')
  plt.plot( RElbow[0], RElbow[1],  marker = 'v', color ="white")
  plt.text( RElbow[0], RElbow[1], 'RElbow[')
  plt.plot( LWrist[0], LWrist[1], marker = 'v', color ="white")
  plt.text( LWrist[0], LWrist[1], 'LWrist')
  plt.plot( RWrist[0], RWrist[1], marker = 'v', color ="white")
  plt.text( RWrist[0], RWrist[1], 'RWrist')
  plt.plot( LHip[0], LHip[1], marker = 'v', color ="white")
  plt.text( LHip[0], LHip[1], 'LHip')
  plt.plot( RHip[0], RHip[1], marker = 'v', color ="white")
  plt.text( RHip[0], RHip[1], 'RHip')
  plt.plot( LKnee[0], LKnee[1], marker = 'v', color ="white")
  plt.text( LKnee[0], LKnee[1], 'LKnee')
  plt.plot( RKnee[0], RKnee[1], marker = 'v', color ="white")
  plt.text( RKnee[0], RKnee[1], 'RKnee')
  plt.plot( LAnkle[0], LAnkle[1], marker = 'v', color ="white")
  plt.text( LAnkle[0], LAnkle[1], 'LAnkle')
  plt.plot( RAnkle[0], RAnkle[1], marker = 'v', color ="white")
  plt.text( RAnkle[0], RAnkle[1], 'RAnkle')


  print(count)
  count = count+1
  plt.imshow(image_data)
  plt.show()
  emotions = ['neutral/unknown']
#sitting standing and walking need to be defined

  sitting = checkifseated(LHip, RHip, LKnee,RKnee,normalization_factor)

  if sitting:
    kneesapart = checkifkneesapart(LKnee, RKnee,LShoulder,RShoulder)
    armsspread = checkifspread(LShoulder,RShoulder, LWrist, RWrist, LElbow, RElbow)
    eyescovered  = handcoverseye(LEye, REye, LWrist, RWrist)
    nosecovered = handscovernose(LWrist,RWrist, Nose)
    handonlobe =  earpull(LEar, REar, LWrist, RWrist)
    spreadelbows = checkspreadelbows(LWrist, RWrist, LElbow, RElbow, LShoulder, RShoulder, normalization_factor)
    emotions.remove('neutral/unknown')
    if eyescovered:
      
      emotions.append('blocking behaviour')
      emotions.append('worry')
    
    if nosecovered:
      emotions.append('shock')
      emotions.append( 'surprise')
      emotions.append( 'fear')
      emotions.append('appreshension')
    
    if handonlobe:
      emotions.append('stressed')
      emotions.append('contemplating/hesitating')
    
    if kneesapart:
      emotions.append('arrogance')
      emotions.append('confidence')
      emotions.append('rude')
    
    if armsspread:
      emotions.append('confidence') 
      emotions.append('ownership') 
      emotions.append('self-assuredness')
    
    if spreadelbows:
      emotions.append('strength')
      emotions.append('confidence')
      emotions.append('self-assuredness')


   
    
    
  if not sitting:
    territoryconfidence = checkstance(LShoulder, RShoulder, LAnkle, RAnkle)
    eyescovered  = handcoverseye(LEye, REye, LWrist, RWrist)
    nosecovered = handscovernose(LWrist,RWrist, Nose)
    handonlobe =  earpull(LEar, REar, LWrist, RWrist)
    hips = hipsout(LHip, RHip, LShoulder, RShoulder)
    emotions.remove('neutral/unknown')
    if eyescovered:
      emotions.append('blocking behaviour')
      emotions.append('worry')
    
    if nosecovered:
      emotions.append('shock')
      emotions.append( 'surprise')
      emotions.append( 'fear')
      emotions.append('appreshension')
    
    if handonlobe:
      emotions.append('stressed')
      emotions.append('contemplating/hesitating')

    if hips:
      emotions.append('inviting notice')
      
    if territoryconfidence:
      emotions.append('confidence')

  print('state of mind of person : ')
  if not emotions:
    emotions.append('neutral/unknown')
  print(emotions)
```
